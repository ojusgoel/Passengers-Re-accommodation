```python
def flights_between(departure_airport,arrival_airport,flight_schedule_df):
    filtered_flights = flight_schedule_df[
        (flight_schedule_df['DepartureAirport'] == departure_airport) &
        (flight_schedule_df['ArrivalAirport'] == arrival_airport)
    ]
    return filtered_flights
```

```python
city_dict =  dict(zip(f['ArrivalAirport'], f['DepartureAirport']))
city_dict #THis is a dictionary with arrival city as key and departure city as value, useful for "back-propagation" when we see lists of connecting flights
```

```python
# Function to get City Mapping
city_mapping = {}

# Iterate over each row in the DataFrame
for index, row in f.iterrows():
    arrival_city = row['ArrivalAirport']
    departure_city = row['DepartureAirport']

    # Check if arrival city is already in the dictionary
    if arrival_city in city_mapping:
        # Append departure city to the existing list
        city_mapping[arrival_city].append(departure_city)
    else:
        # Create a new list for the arrival city
        city_mapping[arrival_city] = [departure_city]
for key in city_mapping.keys():
    city_mapping[key] = np.unique(city_mapping[key])
```

```python
filtered_passengers_list_melted = filtered_passengers_list.melt(id_vars='CUSTOMER_ID', value_vars=['SPECIAL_NAME_CD2', 'SSR_CODE_CD1'], var_name='tag_num', value_name='tags')
merged_passengers_cost = filtered_passengers_list_melted.merge(pd.DataFrame(list(tags_cost.items()), columns=['tags', 'cost']), on='tags', how='left')
total_costs = merged_passengers_cost.groupby('CUSTOMER_ID')['cost'].sum().reset_index()
```
